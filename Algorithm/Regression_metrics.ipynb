# Regression Metrics 

There are different types of metrics as follows:
- Mean Absolute Error
- Mean Square Error
- Root Mean Square Error
- R2 Score
- Adjusted R2 Score

## MAE

The signifcance of absolute is that we wont take the sign of the error, We will just take the error.


### Advantage

1. The error unit is exactly in the unit of the predicted variable.
2. Robust to outliers.

### Disadvantage

1. Modlus function is not differentiable at 0, SO we wont able to differentiate for optimisation.

## Mean Squared Error

### Advantages

1. We can use this as loss function as it is differentiable everywhere.

### Disdavantage

1. Error and predicted variable are not in same unit.
2. Does not work well with outlier. As the square shows the error. Increases a lot incase of outlier. Not Robust to outliers.



## Root Mean Squared Error

### Advantages

1. We can use this as loss function as it is differentiable everywhere.

### Disdavantage

1. Error and predicted variable are not in same unit.
2. Does not work well with outlier. As the square shows the error. Increases a lot incase of outlier. Not Robust to outliers.

## R2 Score

It is also of called coefficient of determination or goodness of fit.
When we calculate R2 score, We basically compare how the linear regression line is better than mean line.

Where SSR is sum of square error in regression line and SSm is sum of squared error in mean line.



### Interpretation of R2 score

**Case 1: When R2 Score is 0**

This happens when the ratio is equal to one, WHich means that the error made by regression line is equal to mean line.

**Case 2: When R2 Score is 1**

WHen the ratio is equal to 0, WHen the numerator is equal to 0. When linear regression line is making no error.

**Case 3: When R2 Score is -ve**
When the numerator is greater than SSm, This means the linear regression line is generating more error than mean.
