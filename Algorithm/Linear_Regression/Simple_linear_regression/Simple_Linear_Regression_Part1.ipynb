# Linear Regression

There are three Types of Linear Regression
- **Simple Linear Regression**
- **Multiple Linear Regression**
- **Polynomial Linear Regression**
- **Regularization**

1. Simple Linear Regression : WHen there is one Input.
2. Multiple Linear Regression : When Input is more than 1.
3. Polynomial Linear Regression : When the relation between is not linear.

## How to anticapte the Package?

We will draw the best fit, the best fit line is the line which does minimum error.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("placement.csv")
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.xlabel("CGPA")
plt.ylabel("package")
```

```python
X = df[['cgpa']]
Y = df[['package']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y,
                                                   test_size = 0.2,
                                                   random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
X_train.shape
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.plot(X_test['cgpa'], y_pred, color = 'red')
plt.xlabel("CGPA")
plt.ylabel("package")
```

```python
# This is the slope
m = lr.coef_

# Intercepy
b = lr.intercept_
```

SO basically, linear regression model is calulcating the slope and intercept.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# MAE
mean_absolute_error(y_test, y_pred)
```

```python
# MSE
mean_squared_error(y_test, y_pred)
```

```python
# RMSE
(mean_squared_error(y_test, y_pred))**0.5
```

```python
# R2 SCORE
r2_score(y_test, y_pred)
```

```python
n = y_test.shape[0]
```

```python
p = X_train.shape[1]
```

```python
n
```

```python
p
```

```python
# Adjusted R2
adj_r2 = 1-(1-r2_score(y_test, y_pred))*(n-1)/(n-p-1)
```
