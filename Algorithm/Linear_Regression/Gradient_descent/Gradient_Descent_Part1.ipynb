# Geometric Intuition

## How do we know in which direction we need to move



### Trick

The Trick is we need to find the slope at the point we have taken.
After calculating the slope at that point by differentiating and plugging in the coordinates value. We perform the following thing.
- +ve slope : We decrease
- -ve slope : We increase



If we subtract/ Add with slope then we will have a drastic change in the b value leading it to a very zig - zag movement. SO we modify the equation by multiplying the learning rate in the slope.
