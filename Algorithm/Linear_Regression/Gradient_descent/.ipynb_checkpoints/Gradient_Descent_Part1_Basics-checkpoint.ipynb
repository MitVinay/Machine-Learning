# Geometric Intuition

## How do we know in which direction we need to move



### Trick

The Trick is we need to find the slope at the point we have taken.
After calculating the slope at that point by differentiating and plugging in the coordinates value. We perform the following thing.
- +ve slope : We decrease
- -ve slope : We increase



If we subtract/ Add with slope then we will have a drastic change in the b value leading it to a very zig - zag movement. SO we modify the equation by multiplying the learning rate in the slope.



# When to Stop

Since its a iterative operation, We need to find when to stop.

### First Approach
- Difference between b-old and b-new is less than 0.0001

### Second Approach
- Fix the number of iteration. The iteration is also called epochs.

# Mathematical Formulation

The given data set below, We need to find the best fit line. That is we need to find m and b. Suppose we know value of m to simply the question. Now We need to iterate with respect of b.



**Step 1**: Start with a random value of b, Say b.
**Step 2**: Iterating with respect to b.




**Step 3**: We need to find the slope. Say slope at b = 0. We need to differentiate the loss function.



**Step 4**:  We use the following formula to calculate b-new.


```python
# Importing library to create a small regression data set.
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X, y = make_regression(n_samples = 4, n_features = 1, n_informative = 1, n_targets = 1, noise = 80, random_state = 13)
```

```python
plt.scatter(X, y)
```

```python
# Intially We will apply Ordinary least Square (OLS) method using Sklearn

reg = LinearRegression()
reg.fit(X, y)
```

```python
# Slope Value after training the model
m = reg.coef_
```

```python
# Intercept value.
reg.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), 'r')
```

```python
# Lets apply Gradient Decsent assuming slope is constant m = 78.35
# and Let's assume the starting value for intercept b = 0
y_pred = (78.35 * X).reshape(4)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), 'r', label = "OLS")
plt.plot(X, y_pred, 'b', label = "Gradient Descent, b = 0")

plt.legend()
```

```python
# iteration 1
b = 0
lr = 0.1

slope = -2 * np.sum(y - m * X.ravel() - b)
```

```python
# stepsize = slope*lr
b_new = b - slope*lr
```

```python
b_new
```

```python
y_pred1 = m*X + b_new
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), 'r', label = "OLS")
plt.plot(X, y_pred, 'b', label = "Gradient Descent, b = 0")
plt.plot(X, y_pred1, 'g', label = "Gradient Descent, b = 20.9277")

plt.legend()
```
