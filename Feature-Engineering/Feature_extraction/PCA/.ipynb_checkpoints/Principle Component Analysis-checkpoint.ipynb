# Principle Component Analysis

PCA is a unsupervised algorithm, There is no output. We only pass input in the algorithm.

PCA is technique for feature extraction. The main objective of this technique is to do dimensionality reduction.
SO PCA trys to achive the best possible lower dimension data to capture the essense of higher dimension data.

## Benefits

- Higher Dimesnion data to lower dimension data, That leads to faster execution of the algorithm.
- Visualisation. For example we have 10 D , this algo will help to convert it to 3D and visualise it.

## Why Variance is so important?

If we see the below data, We can understand that mean is not something which we can use to differentiate between two data set. Moreover mean is only used for central tendency of the data.




So here comes the role of variance, if we see below. The variance of the second data set is more than that of the first data set. However we need to understand that the spread and variance is different.



Spread and Variance are not exactly the same. Although Spread is positively proportional to variance. Accuratly, Spread is Mean Absolute Devaition. In this we use the modulus function intsead of square. The reason why we prefer variance over spreat is modulus function is non- differentiable at 0. Which makes it difficult with PCA algorithm. On the other hand variance is differentiable at every point.

We project the points on the axis and then calculate the variance. That variance is roughly equal to spread.


The end goal is to maximise the variance so that the essense of the data set remains the same. For example give below if we take two points red and green. And project it on x-axis and y-axis. if we compare the projected distance on x-axis and y-axis. The distance on x-axis is still comparable to the actual one. SO incase we are woking on distance based algorith this x-axis distance will give more justice than the y-axis. That why variance is very important.

##### Eigen vectors
Eigen vectors are the vectors on applying the transformation, the direction does not changes.Opposite direction will be considered as eigen vector only becuase intsead of changing direction we say the vector has negative magnitude. i.e direcion remains entact but the magnitude changes.

##### Eigen Values
