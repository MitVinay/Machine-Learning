# Different types of imputation based on number of variable



**There are four ways of imputing values in numerical data**
1. Mean/Median
2. Arbitary
3. End of Distribution
4. Random

## 1. Mean/ Median

1. Incase of Normal distribution , we impute with **mean**.
2. Incase there is any skewness in the distribution, we imput with **median**



**Benefits**

- Very Simple to impute
- Works well with missing data less than 5 percentage. This method doesnot perform well if we have large number of missing data.

**Dis-advantage**

- Changes the distribution of the variable.
- Creates Extra outliers.
- Covariance or correlation changes.

**When to use**

- MCAR
- missing data less than 5 percentage.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('titanic_toy.csv')
```

```python
df.head()
```

```python
# percenatge of missing data
df.isnull().mean()*100
```

```python
X = df.drop('Survived', axis = 1)
Y = df['Survived']
```

```python
# Splitting train and test data
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                   Y,
                                                   test_size = 0.2, 
                                                   random_state = 1)
```

```python
X_train
```

## Imputing Mean/Median with the help of Pandas

```python
# Mean 
mean_age = X_train['Age'].mean()
mean_fare = X_train['Fare'].mean()

median_age = X_train['Age'].median()
median_fare = X_train['Fare'].median()
```

```python
X_train['Age_median'] = X_train['Age'].fillna(median_age)
X_train['Age_mean'] = X_train['Age'].fillna(mean_age)

X_train['fare_median'] = X_train['Fare'].fillna(median_fare)
X_train['fare_mean'] = X_train['Fare'].fillna(mean_fare)
```

```python
print("Age Variance")
print('Original', X_train['Age'].var())
print('Median', X_train['Age_median'].var())
print('Mean', X_train['Age_mean'].var())
print()
print("Fare Variance")
print('Original', X_train['Fare'].var())
print('Median', X_train['fare_median'].var())
print('Mean', X_train['Fare_mean'].var())
```
