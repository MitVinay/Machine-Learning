**Random Value Imputation** and **Missing indicator** are two different algorithm that we use for both categorical and numerical data.



# A. Random Imputation Method

In this method, we simply impute with any random number within the data of that column. In arbitrary we impute with any random variable but with Random imputation we choose the number with in the distribution.

There is no class in sklearn to perform this job. We need to use pandas in order to perform it.

This method works well when we apply linear based algorithm like linear regression, but when we use this method in tree based algorithm this method does not perform well beacuse we are inducing randomness in the data.


**Benefits**

- The distribution of the column does not change.

**Disadvanatges**

- The covariance and correlation changes with other columns.



```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Working on Numerica data

```python
df = pd.read_csv("train_random.csv", usecols = ['Age', 'Fare', 'Survived'])
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
X = df.drop('Survived', axis = 1)
Y = df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                   Y,
                                                   test_size = 0.2,
                                                   random_state = 0)
```

```python
X_train['Age_imputed'] = X_train["Age"]
X_test['Age_imputed'] = X_test["Age"]
```

```python
X_train.head()
```

Now we are imputing 148 values in the age imputed column from the age column.

```python

# Identify missing values in "Age_imputed"
missing_age_indices = X_train[X_train["Age_imputed"].isnull()].index

# Sample non-null values from "Age" column
sampled_age_values = X_train['Age'].dropna().sample(X_train['Age'].isnull().sum()).values

# Assign sampled values to missing indices in "Age_imputed"
X_train.loc[missing_age_indices, 'Age_imputed'] = sampled_age_values
```

```python
# Identify missing values in "Age_imputed"
missing_age_indices1 = X_test[X_test["Age_imputed"].isnull()].index

# Sample non-null values from "Age" column
sampled_age_values1 = X_test['Age'].dropna().sample(X_test['Age'].isnull().sum()).values

# Assign sampled values to missing indices in "Age_imputed"
X_test.loc[missing_age_indices1, 'Age_imputed'] = sampled_age_values1
```

```python
X_train.isnull().sum()
```

```python
X_test.isnull().sum()
```

```python
sns.distplot(X_train['Age'], label = 'Original', hist = False)
sns.distplot(X_train['Age_imputed'], label = 'Imputed', hist = False)

plt.legend()
plt.show()
```

```python
print("Original variance of Age", X_train['Age'].var())
print("New variance of Age", X_train['Age_imputed'].var())
```
