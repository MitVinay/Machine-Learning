```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
# check unique values
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['number'].unique()
```

```python
fig =  df['number'].value_counts().plot.bar()
fig.set_title('Passenger travelling with')
```

```python
# Extract numerical numbers
df['number_numerical'] = pd.to_numeric(df["number"], errors = 'coerce',
                                      downcast = 'integer')
```

```python
df['number_numerical']
```

```python
df['number_categorical'] = np.where(df['number_numerical'].isnull(), df['number'], np.nan)

df.head()
```

```python
# working on cabin column
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')
df['cabin_cat'] = df['Cabin'].str[0]
```

```python
df['cabin_cat']
```

```python
df['cabin_num']
```

```python
df['cabin_cat'].unique()
```

```python
df['cabin_num'].unique()
```

```python
# Extracting ticket details
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split(' ')[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                errors = 'coerce',
                                downcast = 'integer')

# Extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,df['ticket_cat'])
                            
```
