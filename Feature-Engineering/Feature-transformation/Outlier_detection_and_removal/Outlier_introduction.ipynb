# How to Treat Outliers

These are the following ways how we treat outlier in the algorithm
- **Trimming** : In this method we remove the outlier from the data set. If we have lot many outlier, In this approach we will make the data thin i.e. less data. 
- **Capping** : Extreme values are outlier can be in either of the direction. The middle values or avg values cannot be a outlier. So to fix this we put up a capping. And we assign a lower limit values like As shown in the given figure. 1 is assigned to the lower limit and 90 to the upper limit. 
- **Missing Value**: Treating them as missing values
- **Discretization**: Discussed before, creating category from numerical data.
