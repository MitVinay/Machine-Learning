```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# Planning

Imputing missing values ---> OHE --> Scaling --> Feature Selection --> DecisionTree


```python
df.drop(columns = ['PassengerId', 'Ticket', 'Name', 'Ticket'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Splitting data
X_train, X_test, y_train, y_test = train_test_split(df.drop('Survived', axis = 1),
                                                   df['Survived'], 
                                                   test_size = 0.2,
                                                   random_state = 0)
```

There is a reason why the index was given instead of column name in the column transformer. Beacause when you pass something in the column transformer it gives you a array instead of data frame. So the output create in the input of the 2nd column transformer will create a problem as the data passed will be in array and there will be no column name just the index.

```python
# Impute Transformer
trf1 = ColumnTransformer([
    ('impute_age', SimpleImputer(), [2]),
    ('impute_embarked', SimpleImputer(strategy = 'most_frequent') , [6])], remainder = 'passthrough' )
```

**Note:** If we are using feature selection then we need to use minmax scaler instead of standardscaler. If we would not have used feature selection , we can use standardscaler.

```python
# One Hot Encoding
# We did not drop first as we are using Decison tree. Which has no effect multi-colinearity
trf2 = ColumnTransformer([
    ('ohe_sex_embarked', OneHotEncoder(sparse = False, handle_unknown = 'ignore'), [1,6])
], remainder = 'passthrough' )
```

```python
# scaling # Slice helps you to call the column
trf3 = ColumnTransformer([
    ('scale', MinMaxScaler(), slice(0, 10))
])
```

```python
# using top 8 column
# Feature selection 
trf4 = SelectKBest(score_func = chi2, k = 8 )
```

```python
# Train the model
trf5 = DecisionTreeClassifier()
```

# Create Pipeline

```python
# name # object
pipe = Pipeline([
    ('trf1', trf1),
    ('trf2', trf2),
    ('trf3', trf3),
    ('trf4', trf4),
    ('trf5', trf5),
])
```

# Pipeline Vs make_pipeline

Pipeline requires naming of steps, make_pipeline does not.

( Same applies to Column Transformer vs make_column_transformer)

```python
# Alternate syntax
pipe = make_pipeline(trf1, trf2, trf3, trf4, trf5)
```

```python
pipe.fit(X_train, y_train)
```

There is major difference
- First if you are not calling a model in the pipeline, then you are just doing feature transformation. Then either you can first **fit** then **transform** or directly use fit_transform.
- Secondly if you are inserting a model then just only use  **fit** function then you can use **predict** function.
